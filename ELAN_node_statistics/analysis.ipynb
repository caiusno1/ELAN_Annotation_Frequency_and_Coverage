```python
import glob     # Import glob to easily loop over files
import pympi    # Import pympi to work with elan files
import nltk     # Import nltk to use tokenizer
import pandas as pd
```

```python
keys = pd.read_csv("Nodeaufstellung.csv")["NodeID"].to_list()
files = glob.glob("in/*.eaf")
df = pd.DataFrame({"file":[],"role":[],"VP":[],"node":[],"frequency":[]})
dfCoverage = pd.DataFrame({"file":[],"role":[],"VP":[],"coverage":[]})
errorlog = ""
for file in files:
    eafob = pympi.Elan.Eaf(file)
    singleTextEX = "( EX "+file+")"
    singleTextEE = "( EE "+file+")"
    for annotation in eafob.get_annotation_data_for_tier("EX_Gamenodes"):
        singleTextEX=singleTextEX+","+annotation[2]
    for annotation in eafob.get_annotation_data_for_tier("EE_Gamenodes"):
        singleTextEE=singleTextEE+","+annotation[2]

    EXtokens = list(map(lambda x: x.strip(), singleTextEX.split(",")))
    EEtokens = list(map(lambda x: x.strip(), singleTextEE.split(",")))
    EXtokenCount = {}
    for token in EXtokens:
        if not token in keys and not token =="":
            token2 = token+"."
            token3 = token.replace("X","").strip()
            token4 = token.replace("x","").strip()
            token5 = token.replace("x","").replace("X","").strip()+"."
            if token2 in keys:
                errorlog = errorlog + "Node name corrected: "+token+"->"+token2+"\n"
                print("Node name corrected: "+token+"->"+token2)
                token = token2
            elif token3 in keys:
                errorlog = errorlog + "Node name corrected: "+token+"->"+token3+"\n"
                print("Node name corrected: "+token+"->"+token3)
                token = token3
            elif token4 in keys:
                errorlog = errorlog + "Node name corrected: "+token+"->"+token4+"\n"
                print("Node name corrected: "+token+"->"+token4)
                token = token4
            elif token5 in keys:
                errorlog = errorlog + "Node name corrected: "+token+"->"+token5+"\n"
                print("Node name corrected: "+token+"->"+token5)
                token = token5
            elif token == "1.0":
                errorlog = errorlog + "Node name corrected: "+token+"->1\n"
                print("Node name corrected: "+token+"->1")
                token = "1"
            elif token == "1.0.":
                errorlog = errorlog + "Node name corrected: "+token+"->1\n"
                print("Node name corrected: "+token+"->1")
                token = "1"
            elif token == "1.":
                errorlog = errorlog + "Node name corrected: "+token+"->1\n"
                print("Node name corrected: "+token+"->1")
                token = "1"
            elif token == "8.2.1":
                errorlog = errorlog + "Node name corrected: "+token+"->8.2\n"
                print("Node name corrected: "+token+"->8.2")
                token = "8.2"
            elif token == "5.7.":
                errorlog = errorlog + "Node name corrected: "+token+"->6.2.2.\n"
                print("Node name corrected: "+token+"->6.2.2.")
                token = "6.2.2."
            elif token.startswith("(") and token.endswith(")"):
                errorlog = errorlog + ""+token[1:-1]+"\n"
                continue
            else:
                errorlog = errorlog + "Node unknown: "+token+"\n"
                print("Node unknown: "+token)
                continue
        if token in EXtokenCount.keys():
            EXtokenCount[token] = EXtokenCount[token] + 1
        elif not token.startswith("(") or token.endswith(")"):
            EXtokenCount[token] =  1
    for nodeName in keys:
        if not nodeName in EXtokenCount.keys():
            EXtokenCount[nodeName]=0
    #EXtokenCount.pop("")
    EEtokenCount = {}
    for token in EEtokens:
        if not token in keys and not token =="":
            token2 = token+"."
            token3 = token.replace("X","").strip()
            token4 = token.replace("x","").strip()
            token5 = token.replace("x","").replace("X","").strip()+"."
            if token2 in keys:
                errorlog = errorlog + "Node name corrected: "+token+"->"+token2+"\n"
                print("Node name corrected: "+token+"->"+token2)
                token = token2
            elif token3 in keys:
                errorlog = errorlog + "Node name corrected: "+token+"->"+token3+"\n"
                print("Node name corrected: "+token+"->"+token3)
                token = token3
            elif token4 in keys:
                errorlog = errorlog + "Node name corrected: "+token+"->"+token4+"\n"
                print("Node name corrected: "+token+"->"+token4)
                token = token4
            elif token5 in keys:
                errorlog = errorlog + "Node name corrected: "+token+"->"+token5+"\n"
                print("Node name corrected: "+token+"->"+token5)
                token = token5
            elif token == "1.0":
                errorlog = errorlog + "Node name corrected: "+token+"->1\n"
                print("Node name corrected: "+token+"->1")
                token = "1"
            elif token == "1.0.":
                errorlog = errorlog + "Node name corrected: "+token+"->1\n"
                print("Node name corrected: "+token+"->1")
                token = "1"
            elif token == "1.":
                errorlog = errorlog + "Node name corrected: "+token+"->1\n"
                print("Node name corrected: "+token+"->1")
                token = "1"
            elif token == "8.2.1":
                errorlog = errorlog + "Node name corrected: "+token+"->8.2\n"
                print("Node name corrected: "+token+"->8.2")
                token = "8.2"
            elif token == "5.7.":
                errorlog = errorlog + "Node name corrected: "+token+"->6.2.2.\n"
                print("Node name corrected: "+token+"->6.2.2.")
                token = "6.2.2."
            elif token.startswith("(") and token.endswith(")"):
                errorlog = errorlog + ""+token[1:-1]+"\n"
                continue
            else:
                errorlog = errorlog + "Node unknown: "+token+"\n"
                print("Node unknown: "+token)
                continue
        if token in EEtokenCount.keys():
            EEtokenCount[token] = EEtokenCount[token] + 1
        elif not token.startswith("(") or token.endswith(")"):
            EEtokenCount[token] =  1
    for nodeName in keys:
        if not nodeName in EEtokenCount.keys():
            EEtokenCount[nodeName]=0
    #EEtokenCount.pop("")
    for nodeName in keys:
        df = pd.concat([df, pd.DataFrame({"file":[file],"role":["EE"],"VP":[file.replace("in\\","").replace("_P2_A01.eaf","")],"node":[nodeName],"frequency":[EEtokenCount[nodeName]]})])
        df = pd.concat([df, pd.DataFrame({"file":[file],"role":["EX"],"VP":[file.replace("in\\","").replace("_P2_A01.eaf","")],"node":[nodeName],"frequency":[EXtokenCount[nodeName]]})])
    totalNodes = len(keys)
    EECoveragedNodes = len(list(filter(lambda x: x != 0,EEtokenCount.values())))
    EXCoveragedNodes = len(list(filter(lambda x: x != 0,EXtokenCount.values())))
    dfCoverage = pd.concat([dfCoverage, pd.DataFrame({"file":[file],"role":["EE"],"VP":[file.replace("in\\","").replace("_P2_A01.eaf","")],"coverage":[EECoveragedNodes/totalNodes]})])
    dfCoverage = pd.concat([dfCoverage, pd.DataFrame({"file":[file],"role":["EX"],"VP":[file.replace("in\\","").replace("_P2_A01.eaf","")],"coverage":[EXCoveragedNodes/totalNodes]})])
df.to_csv("out/frequencies.csv", sep=";", index=False)
dfCoverage.to_csv("out/coverage.csv", sep=";", index=False)
with open("out/error.log","w") as f:
    f.write(errorlog)
```
